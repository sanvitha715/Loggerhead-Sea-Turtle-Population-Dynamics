# Loggerhead Sea Turtle Population Dynamics — Figures

This notebook reproduces the three figures from the project report:

1. **Structure of the 7×7 projection matrix**
2. **Summary of the seven life stages**
3. **Comparison of dominant eigenvalues** (baseline vs. improved survival)

Run all cells top to bottom. Requires `numpy` and `matplotlib`.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Figure 1 — Structure of the 7×7 Sea Turtle Projection Matrix

The first row holds fecundity values, the diagonal represents remaining in the same stage, and the subdiagonal represents advancing to the next stage.

```python
# The 7x7 sea turtle projection matrix
A = np.array([
    [0, 0, 0, 0, 127, 4, 80],
    [0.6747, 0.7370, 0, 0, 0, 0, 0],
    [0, 0.0486, 0.6610, 0, 0, 0, 0],
    [0, 0, 0.0147, 0.6907, 0, 0, 0],
    [0, 0, 0, 0.0518, 0, 0, 0],
    [0, 0, 0, 0, 0.8091, 0, 0],
    [0, 0, 0, 0, 0, 0.8091, 0.8089],
])

fig, ax = plt.subplots(figsize=(7, 6))

# Shade the nonzero entries so the structure is easy to see
ax.imshow(A != 0, cmap="Greys", interpolation="none")

# Write each value inside its cell
for i in range(7):
    for j in range(7):
        ax.text(j, i, A[i, j], ha="center", va="center", fontsize=9)

ax.set_xticks(range(7))
ax.set_yticks(range(7))
ax.set_xticklabels([1, 2, 3, 4, 5, 6, 7])
ax.set_yticklabels([1, 2, 3, 4, 5, 6, 7])
ax.set_xlabel("Stage at time t")
ax.set_ylabel("Stage at time t+1")
ax.set_title("Figure 1. Structure of the 7x7 Sea Turtle Projection Matrix")

ax.text(5, -1.2, "First row = fecundity", ha="center", fontsize=10)
ax.text(5.8, 6.9, "Diagonal = stay in same stage", ha="left", fontsize=9)
ax.text(1.2, 4.8, "Subdiagonal = move to next stage", ha="left", fontsize=9)

plt.tight_layout()
plt.show()
```

## Figure 2 — Summary of the Seven Stages

Biological role of each of the seven life stages used in the model.

```python
data = [
    ["1", "Hatchlings", "Early life stage"],
    ["2", "Small Juveniles", "Growth and survival"],
    ["3", "Large Juveniles", "Growth and survival"],
    ["4", "Subadults", "Transition toward breeding"],
    ["5", "Novice Breeders", "First breeding stage"],
    ["6", "1st-Year Remigrants", "Post-breeding stage"],
    ["7", "Mature Breeders", "Adult breeding stage"],
]

columns = ["Stage", "Name", "Role in model"]

fig, ax = plt.subplots(figsize=(9, 4))
ax.axis("off")

table = ax.table(
    cellText=data,
    colLabels=columns,
    loc="center",
    cellLoc="center",
)
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.5)

plt.title("Figure 2. Summary of the Seven Stages in the Sea Turtle Model", pad=12)
plt.show()
```

## Figure 3 — Comparison of Dominant Eigenvalues

Baseline model (λ ≈ 0.945, decline) versus the improved-survival scenario (λ ≈ 1.02, growth). The dashed line at 1.0 marks the growth/decline cutoff.

```python
labels = ["Baseline model", "Improved survival\nscenario"]
values = [0.945, 1.02]

plt.figure(figsize=(6, 4))
plt.bar(labels, values)

# Dashed line at 1 marks the growth/decline cutoff
plt.axhline(1.0, linestyle="--", linewidth=1)

for i, v in enumerate(values):
    plt.text(i, v + 0.01, str(v), ha="center")

plt.ylabel("Dominant eigenvalue")
plt.title("Figure 3. Comparison of Dominant Eigenvalues")
plt.show()
```
